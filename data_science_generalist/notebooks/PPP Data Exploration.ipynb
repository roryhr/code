```python
big_df[big_df.JobsRetained > 0][['LoanRange', 'JobsRetained']].groupby('LoanRange').mean()
```

```python
(
    big_df[big_df.JobsRetained > 0][
    ['LoanRange', 'JobsRetained']
]
    .groupby('LoanRange')
    .mean()
    .plot
    .bar()
)
```

```python
jobs = (
    big_df[big_df.JobsRetained > 0][
    ['LoanRange', 'JobsRetained']
]
    .groupby('LoanRange')
    .agg({'mean', 'count', 'sum'})
    .reset_index()
)
jobs
```

```python
jobs['MiddleLoan'] = [7.5e6, 3.5e6, 1.5e6, (.35 + 1)/2*1e6, .25e6]
```

```python
jobs
```

```python
jobs['Loan per job'] = jobs['MiddleLoan'] * jobs['JobsRetained']['count']/jobs['JobsRetained']['sum']
```
