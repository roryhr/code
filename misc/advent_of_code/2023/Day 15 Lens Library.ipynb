```python
puzzle_input = """
rn=1,cm-,qp=3,cm=2,qp-,pc=4,ot=9,ab=5,pc-,pc=6,ot=7
"""
```

```python
steps = puzzle_input.strip().split(",")
steps
```

```python
def hash_algorithm(string):
    current_value = 0
    
    for char in string:
        ascii_code = ord(char)

        current_value += ascii_code
        current_value *= 17
        current_value %= 256
    
    return current_value

hash_algorithm("HASH")
```

```python
sum(hash_algorithm(step) for step in steps)
```
