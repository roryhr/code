# Example

```python
puzzle_input = """
1000
2000
3000

4000

5000
6000

7000
8000
9000

10000
"""
```

```python
unsplit = puzzle_input.strip().split('\n\n')
unsplit
```

```python
calories = [
    x.split()
    for x in unsplit
]

calories
```

```python
calories = [
    sum(int(x) for x in bag) for bag in calories
]
```

```python
calories
```

```python
max(calories)
```
