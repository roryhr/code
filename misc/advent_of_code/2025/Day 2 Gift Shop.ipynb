# Example


```python
puzzle_input = """
11-22,95-115,998-1012,1188511880-1188511890,222220-222224,
1698522-1698528,446443-446449,38593856-38593862,565653-565659,
824824821-824824827,2121212118-2121212124
"""
```

```python
split = puzzle_input.strip().split(",")
split
```

```python
[s.split("-") for s in split][0]
```

```python
def check_substrings(word):
    half_word_length = len(word) // 2
    return word[:half_word_length] == word[half_word_length:]
```

```python
check_substrings("1111")
```

```python
invalid_ids = []

for s in split:
    start, end = s.strip().split("-")
    start = int(start)
    end = int(end)

    for i in range(start, end + 1):
        if check_substrings(str(i)):
            invalid_ids.append(i)
```

```python
invalid_ids
```

```python
sum(invalid_ids)
```

# Part 2

```python
def check_part2(word):
    # Take all slices
    word_length = len(word)

    for slice_length in range(1, word_length // 2 + 1):
        chunks = [
            word[i : i + slice_length] for i in range(0, word_length, slice_length)
        ]
        # print(chunks)
        if len(set(chunks)) == 1:
            return True


print(check_part2("111"))
```
