```python
puzzle_input = """
987654321111111
811111111111119
234234234234278
818181911112111
"""
```

```python
banks = puzzle_input.strip().split("\n")
banks
```

```python
def max_joltage(numbers_str):
    numbers = [int(n) for n in numbers_str]
    pairs = []
    # For each num take it and the and max of everything after
    for i, n in enumerate(numbers[:-1]):
        pairs.append((n, max(numbers[i + 1 :])))

    return max([int(str(i) + str(j)) for i, j in pairs])


max_joltage("987654321111112")
```

```python
max_joltage("811111111111119")
```

```python
joltages = [max_joltage(bank) for bank in banks]
joltages
```

```python
sum(joltages)
```

```python
def get_biggest(bank, amount):
    number = ""
    for x in range(-amount + 1, 0):
        digit = str(max(map(int, bank[:x])))
        pos = bank.find(digit)
        number = number + digit
        bank = bank[pos + 1 :]
    number = number + max(bank)
    return int(number)


get_biggest("811111111111119", 2)
```

```python
joltages = [get_biggest(bank, 12) for bank in banks]
```

```python
sum(joltages)
```
