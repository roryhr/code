```python
class Tile:
    def __init__(self, x, y, is_wall):
        self.x = x
        self.y = y
        self.is_wall = is_wall

    def __repr__(self):
        return f"Tile({self.x}, {self.y})"


def get_neighbors(maze, tile):
    x, y = tile.x, tile.y
    neighbors = []
    for dx, dy in [(0, 1), (1, 0), (0, -1), (-1, 0)]:  # Right, Down, Left, Up
        nx, ny = x + dx, y + dy
        if 0 <= nx < len(maze[0]) and 0 <= ny < len(maze):
            neighbor_tile = maze[ny][nx]
            if not neighbor_tile.is_wall:
                neighbors.append(neighbor_tile)
    return neighbors


def bfs_shortest_path(maze, start_tile, end_tile):
    queue = deque([(start_tile, 0)])  # (Tile, Score)
    visited = set([start_tile])

    while queue:
        current_tile, score = queue.popleft()
        if current_tile == end_tile:
            return score

        for neighbor in get_neighbors(maze, current_tile):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(
                    (neighbor, score + 1)
                )  # Increment score for moving forward

    return -1  # No path found


# Example Usage (replace with your actual maze data)
maze = [[Tile(x, y, False) for y in range(len(row))] for x in range(len(rows))]
```
