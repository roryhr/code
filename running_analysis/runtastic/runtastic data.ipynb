# Sessions

```python
p = Path('/Users/rory/data/runtastic/export-20191017-000/Sport-sessions/')
```

```python
series_list = [
    pd.read_json(json_file, typ='series')
    for json_file
    in p.glob('*.json')
]
```

```python
sessions = pd.DataFrame(series_list)
```

```python
import mplleaflet
```

```python
sessions[sessions.longitude.between(13, 15)].plot.scatter(x='longitude', y='latitude')
mplleaflet.display()
```

```python
session_stats = sessions[sessions.longitude > 0].sum()
session_stats
```

```python
session_stats.duration / 1000 / 3600 # ms -> sec -> hours
```

```python
session_stats.distance / 1000  # m -> km
```
