I'll update this graph as I try different beers!
The graph ordering is random so I rerun this until I get a layout I like. 


```python
from functools import reduce

import matplotlib.pyplot as plt
import networkx as nx
```

```python
G = nx.DiGraph()

edges = [
    ("Half Acre", "Beyond the Mountain"),
    ("Beyond the Mountain", "Breckenridge Brewery"),
    ("Breckenridge Brewery", "Hofbräu"),
    # ("Yuengling", "Pabst Blue Ribbon"),
    # ("Pabst Blue Ribbon", "Budweiser"),
]
combined = reduce(lambda acc, x: acc + list(x), edges, [])
G.add_nodes_from(set(combined))

G.add_edges_from(edges)

# Draw the graph with beer labels
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Use spring layout for a clean look
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="lightblue",
    node_size=3000,
    font_size=10,
    arrows=True,
    font_weight="bold",
)

plt.title("Graph of Oktoberfest Beers")
plt.show()
```
